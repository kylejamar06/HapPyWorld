```python
# Importing Dependencies

import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import & compare file heads of our various CSVs. We'll use 2019.csv as our template and create new
# dataframes of off the rest to ensure the data is congruent.

csvpath_2015 = os.path.join('Resources', '2015.csv')
csvpath_2016 = os.path.join('Resources', '2016.csv')
csvpath_2017 = os.path.join('Resources', '2017.csv')
csvpath_2018 = os.path.join('Resources', '2018.csv')
csvpath_2019 = os.path.join('Resources', '2019.csv')

csv_2015 = pd.read_csv(csvpath_2015)
csv_2015.head()
```

```python
csv_2016 = pd.read_csv(csvpath_2016)
csv_2016.head()
```

```python
csv_2017 = pd.read_csv(csvpath_2017)
csv_2017.head()
```

```python
csv_2018 = pd.read_csv(csvpath_2018)
csv_2018.head()
```

```python
csv_2019 = pd.read_csv(csvpath_2019)
csv_2019.head()
```

```python
csv_2017.head()
```

```python
# 2019 & 2018 datasets are the same in structure & column naming convention.  Restructure / Rename
# datasets from 2015 - 2017.

# Convert / Fix 2017 Data

csv_2017_restructured = csv_2017[["Happiness.Rank",
                                  "Country",
                                  "Happiness.Score",
                                  "Economy..GDP.per.Capita.",
                                  "Family",
                                  "Health..Life.Expectancy.",
                                  "Freedom",
                                  "Generosity",
                                  "Trust..Government.Corruption."]]

csv_2017_restructured = csv_2017_restructured.rename(columns={"Happiness.Rank": "Overall rank",
                                      "Country": "Country or region",
                                      "Happiness.Score": "2017 Score",
                                      "Economy..GDP.per.Capita.": "GDP per capita",
                                      "Family": "Social support",
                                      "Health..Life.Expectancy.": "Healthy life expectancy",
                                      "Freedom": "Freedom to make life choices",
                                      "Generosity": "Generosity",
                                      "Trust..Government.Corruption.": "Perceptions of corruption"})
csv_2017_restructured.head()
```

```python
# 2019 & 2018 datasets are the same in structure & column naming convention.  Restructure / Rename
# datasets from 2015 - 2017.

# Convert / fix 2016 Data

csv_2016_restructured = csv_2016[["Happiness Rank",
                                 "Country",
                                  "Happiness Score",
                                  "Economy (GDP per Capita)",
                                  "Family",
                                  "Health (Life Expectancy)",
                                  "Freedom",
                                  "Generosity",
                                  "Trust (Government Corruption)"]]

csv_2016_restructured = csv_2016_restructured.rename(columns={"Happiness Rank": "Overall rank",
                                      "Country": "Country or region",
                                      "Happiness Score": "2016 Score",
                                      "Economy (GDP per Capita)": "GDP per capita",
                                      "Family": "Social support",
                                      "Health (Life Expectancy)": "Healthy life expectancy",
                                      "Freedom": "Freedom to make life choices",
                                      "Generosity": "Generosity",
                                      "Trust (Government Corruption)": "Perceptions of corruption"})
csv_2016_restructured.head()
```

```python
# 2019 & 2018 datasets are the same in structure & column naming convention.  Restructure / Rename
# datasets from 2015 - 2017.

# Convert / fix 2015 Data

csv_2015_restructured = csv_2015[["Happiness Rank",
                                 "Country",
                                  "Happiness Score",
                                  "Economy (GDP per Capita)",
                                  "Family",
                                  "Health (Life Expectancy)",
                                  "Freedom",
                                  "Generosity",
                                  "Trust (Government Corruption)"]]

csv_2015_restructured = csv_2015_restructured.rename(columns={"Happiness Rank": "2015 Overall rank",
                                      "Country": "Country or region",
                                      "Happiness Score": "2015 Score",
                                      "Economy (GDP per Capita)": "2015 GDP per capita",
                                      "Family": "2015 Social support",
                                      "Health (Life Expectancy)": "2015 Healthy life expectancy",
                                      "Freedom": "2015 Freedom to make life choices",
                                      "Generosity": "2015 Generosity",
                                      "Trust (Government Corruption)": "2015 Perceptions of corruption"})
csv_2015_restructured.head()
```

```python
#Rename columns for 2019 to allow these to stay separate when merging.
csv_2019 = csv_2019.rename(columns={"Overall rank": "2019 Overall rank",
                                      "Country or region": "Country or region",
                                      "Score": "2019 Score",
                                      "GDP per capita": "2019 GDP per capita",
                                      "Social support": "2019 Social support",
                                      "Healthy life expectancy": "2019 Healthy life expectancy",
                                      "Freedom to make life choices": "2019 Freedom to make life choices",
                                      "Generosity": "2019 Generosity",
                                      "Perceptions of corruption": "2019 Perceptions of corruption"})
csv_2019.head()
```

```python
merged_data = csv_2015_restructured.merge(csv_2019, on='Country or region')
merged_data['happiness_change'] = (merged_data['2019 Score'] - merged_data['2015 Score']).round(2)
merged_data['gdp_change'] = (merged_data['2019 GDP per capita'] - merged_data['2015 GDP per capita']).round(2)

merged_data = merged_data.sort_values(by=['happiness_change'])
top_decreases = merged_data.nsmallest(5, 'happiness_change', keep='first')

top_increases = merged_data.nlargest(5, 'happiness_change', keep='first')

top_inc_dec = top_decreases.append(top_increases)
top_inc_dec

x_axis = top_inc_dec['Country or region']
y_axis = top_inc_dec['happiness_change']
plt.figure(figsize=(8,4))
plt.bar(x_axis, y_axis, color=['black', 'red', 'green', 'blue', 'cyan'], alpha=0.5)
plt.xticks(rotation=90)
plt.title('Top 5 Increases/Decreases in Happiness (2015-2019)') 
plt.xlabel('Country')
plt.ylabel('Change in Happiness')
plt.savefig("happiness_change.jpg")
```

```python
x_axis = top_inc_dec['Country or region']
y_axis = top_inc_dec['gdp_change']
plt.figure(figsize=(8,4))
plt.bar(x_axis, y_axis, color=['black', 'red', 'green', 'blue', 'cyan'], alpha=0.5, align="center")
plt.xticks(rotation=90)
plt.title('Change in GDP for Top 5 Increases/Decreases in Happiness (2015-2019)') 
plt.xlabel('Country')
plt.ylabel('Change in GDP')
plt.savefig("gdp_change.jpg")
```
